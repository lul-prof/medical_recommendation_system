```python
import pandas as pd
kidney_data=pd.read_csv('CKD_Preprocessed.csv')
```

```python
x=kidney_data.drop("Chronic Kidney Disease: yes", axis=1)
```

```python
x
```

```python
y=kidney_data["Chronic Kidney Disease: yes"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
le.fit(y)
Y=le.transform(y)
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,Y, test_size=0.3, random_state=20)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.datasets import make_classification
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import accuracy_score, confusion_matrix
import numpy as np
```

```python
#Dictionary to store our models
models={
    "SVC":SVC(kernel='linear'),
    "RandomForest":RandomForestClassifier(n_estimators=100,random_state=42),
    "GradientBoosting":GradientBoostingClassifier(n_estimators=100,random_state=42),
    "KNeighbors":KNeighborsClassifier(n_neighbors=5),
    "MultinomialNB":MultinomialNB()
}
for model_name, model in models.items():
    #train model
    model.fit(X_train,y_train)

    #test model
    predictions=model.predict(X_test)

    #xalculate accuracy

    accuracy=accuracy_score(y_test,predictions)

    #calculate confusion matrix

    cm=confusion_matrix(y_test, predictions)

    print(f"{model_name} accuracy:{accuracy}")
    print(f"{model_name} confusion_matrix:")
    print(np.array2string(cm,separator=','))
```

```python
svc=SVC(kernel='linear')
svc.fit(X_train,y_train)
ypred=svc.predict(X_test)
accuracy_score(y_test,ypred)
```

```python
import pickle
pickle.dump(svc,open("kidney.pkl",'wb'))
```

```python
#load the model
svc=pickle.load(open("kidney.pkl",'rb'))
svc
```
