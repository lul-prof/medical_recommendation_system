```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
heart_data=pd.read_csv('heart_disease_data.csv')
```

```python
heart_data.head(20)
```

```python
heart_data.tail(20)
```

```python
heart_data.describe()
```

```python
heart_data.shape
```

```python
heart_data.info()
```

```python
#Checking for missing values
heart_data.isnull().sum()
```

```python
heart_data['target'].value_counts()
```

```python
#splitting data

X=heart_data.drop(columns='target', axis=1)
Y=heart_data['target']
```

```python
X
```

```python
Y
```

```python
scaler=StandardScaler()
scaler.fit(X)
```

```python
standardized_data=scaler.transform(X)
```

```python
standardized_data
```

```python
X=standardized_data
Y=heart_data['target']
```

```python
#Splitting data into trainig and test data
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

```python
#Model Training Via Logistic Regression
model=LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
#Model Evaluation
```

```python
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction, Y_train)
```

```python
print("Training Data Accuracy::",training_data_accuracy )
```

```python
#Test Data Accuracy
```

```python
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction, Y_test)
```

```python
print("Test Data Accuracy::",test_data_accuracy)
```

```python
#Predicting with Known Values
```

```python
input_data=(51,1,0,140,298,0,1,122,1,4.2,1,3,3)
#change the input data to a numpy array
input_data_as_numpy_array=np.asarray(input_data)

#reshape the numpy array coz we are predicting for one instance
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)

prediction=model.predict(input_data_reshaped)

print(prediction)

if (prediction[0]==0):
    print("Heart Is Healthy")
else:
    print("Heart Defect Detected")
    
```
